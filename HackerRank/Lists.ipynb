````python
#MyCode

    N = int(input())
    lists = []
    for _ in range(N):
        comm = list(map(str,input().split()))
        for str_ in comm:
            if 'insert' in str_:
                num1 = int(comm[1])
                num2 = int(comm[2])
                lists.insert(num1,num2)
            elif 'append' in str_:
                num1 = int(comm[1])
                lists.append(num1)
            elif 'remove' in str_:
                num1 = int(comm[1])
                lists.remove(num1)
            elif 'sort' in str_:
                lists.sort()
            elif 'reverse' in str_:
                lists.reverse()
            elif 'pop' in str_:
                lists.pop()
            elif 'print' in str_:
                print(lists)
````

````python
# OtherCode

T = int(input())
L = []
for t in range(T):
    args = input().strip().split(" ")
    if args[0] == "append":
        L.append(int(args[1]))
    elif args[0] == "insert":
        L.insert(int(args[1]), int(args[2]))
    elif args[0] == "remove":
        L.remove(int(args[1]))
    elif args[0] == "pop":
        L.pop()
    elif args[0] == "sort":
        L.sort()
    elif args[0] == "reverse":
        L.reverse()
    elif args[0] == "print":
        print(L)
````

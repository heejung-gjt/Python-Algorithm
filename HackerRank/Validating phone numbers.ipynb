````python
import re
num = int(input())

for _ in range(num):
    number = input()
    if re.match('[7-9][0-9]{9}$',number):
        print('YES')
    else:
        print('NO')  
````

````python
import re
num = int(input())

for _ in range(num):
    number = input()
    if re.fullmatch('[7-9][0-9]{9}',number):
        print('YES')
    else:
        print('NO')  
````

#### í•´ë‹¹ ì •ê·œí‘œí˜„ì‹ì„ ë¨¼ì € ì»´íŒŒì¼ í•´ì£¼ê³  fullmatchë¡œ ê²€ì‚¬í•œë‹¤

````python
n = int(input())
phonenum_pattern = re.compile(r'[7-9][0-9]{9}')
for _ in range(n):
    phonenum_string = input()
    result = phonenum_pattern.fullmatch(phonenum_string)
    if result:
        print('YES')
    else:
        print('NO')
````
